### (opt) cnn regression

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Concatenate

# Load CSV files
good_df = pd.read_csv('../outfits/good4_outfits.csv')

# Add labels
good_df['label'] = 0

def modify_label(row):
    style_number = int(row['index'].split('_')[0])  # 取出 index 前綴的 style number
    if flag == style_number:
        return 1
    else:
        return 0

flag = 0  # 這裡可以調整 flag 的值
good_df['label'] = good_df.apply(modify_label, axis=1) # modify_label or modify_label_cnn
display(good_df)

# Combine the datasets
data = good_df[['index', 'img_pathE', 'img_pathQ', 'label']]

                  
# Function to load and preprocess images
def load_and_preprocess_image(img_path, target_size=(224, 224)):
    img = load_img(img_path, target_size=target_size)
    img = img_to_array(img)
    img = img / 255.0  # Normalize
    return img

# Create a dataset
X1, X2, y = [], [], []

for _, row in data.iterrows():
    imgE = load_and_preprocess_image(row['img_pathE'])
    imgQ = load_and_preprocess_image(row['img_pathQ'])
    X1.append(imgE)
    X2.append(imgQ)
    y.append(row['label'])

X1 = np.array(X1)
X2 = np.array(X2)
y = np.array(y)

# CNN model for regression
inputE = Input(shape=(224, 224, 3))
inputQ = Input(shape=(224, 224, 3))

# Define CNN for imgE
x1 = Conv2D(32, (3, 3), activation='relu')(inputE)
x1 = MaxPooling2D((2, 2))(x1)
x1 = Flatten()(x1)

# Define CNN for imgQ
x2 = Conv2D(32, (3, 3), activation='relu')(inputQ)
x2 = MaxPooling2D((2, 2))(x2)
x2 = Flatten()(x2)

# Concatenate the outputs
combined = Concatenate()([x1, x2])

# Dense layers
x = Dense(64, activation='relu')(combined)
output = Dense(1, activation='sigmoid')(x)

# Create model
model = Model(inputs=[inputE, inputQ], outputs=output)

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', 'mean_squared_error'])
#model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

model.summary()
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets
X1_train, X1_test, X2_train, X2_test, y_train, y_test = train_test_split(
    X1, X2, y, test_size=0.1, random_state=42, stratify = y)

# Check the shapes to confirm the split
print(f"Train shapes: X1: {X1_train.shape}, X2: {X2_train.shape}, y: {y_train.shape}")
print(f"Test shapes: X1: {X1_test.shape}, X2: {X2_test.shape}, y: {y_test.shape}")
```

```python
# Train the model
model.fit([X1_train, X2_train], y_train, epochs=1, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model on the test set
test_loss, test_acc, test_mse = model.evaluate([X1_test, X2_test], y_test)
print(f"Test loss: {test_loss:.4f}")
print(f"Test accuracy: {test_acc:.4f}")
print(f"Test mse: {test_mse:.4f}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

y_pred = model.predict([X1_test, X2_test])
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'[SVM] MSE = {mse}, MAE = {mae}')
```
