### img --> embedding

```python
# im-project-main/models
MODEL_PATH = "../embeddings/lib/df1.pkl"
CLASSES_PATH = "../embeddings/lib/attribute-classes.txt"
```

```python
import torch
from torchvision import transforms
from torch.autograd import Variable
from PIL import Image
import torch.nn as nn
import os


class AttributePredictionModel:
    
    def __init__(self):
        self.model = None
        self.labels = []
        
    def load(self, model_path, labels_path, eval_mode=False):
        self.model = torch.load(model_path)
        self.model.eval()  # mode: eval
        self.labels = open(labels_path, 'r').read().splitlines()
        
        if eval_mode:
            print(self.model)
    
    def predict(self, img, threshold=0.1):
        device = torch.device("cpu")
        #img = Image.open(image_path).convert('RGB')
        
        test_transforms = transforms.Compose([
            transforms.Resize(224),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])
        
        image_tensor = test_transforms(img).float()
        image_tensor = image_tensor.unsqueeze_(0)  # 增加 batch 維度
        inp = Variable(image_tensor).to(device)
        
        with torch.no_grad():
            output = self.model(inp)
        
        # raw output --> prob[0, 1], pred{0, 1}, att
        probabilities = torch.sigmoid(output).cpu().numpy()[0]
        predictions = (probabilities >= threshold).astype(int)
        predicted_attributes = [self.labels[i] for i in range(len(predictions)) if predictions[i] == 1]
        return probabilities, predictions, predicted_attributes

# model instance
learner = AttributePredictionModel()
learner.load(MODEL_PATH, CLASSES_PATH)
```

```python
import pandas as pd
from PIL import Image
import numpy as np

def path2img(csv_path):
    df = pd.read_csv(csv_path)
    imgEs = []

    for index, row in df.iterrows():
        image_path = row['img_path']
        img = Image.open(image_path).convert('RGB')
        img = img.resize((224, 224))
        img_array = np.array(img) # img --> arr
        imgEs.append(img_array)

    imgEs = np.array(imgEs)
    return imgEs

csvE = 'test_imgE.csv'
imgEs = path2img(csvE)
print("imgEs:", imgEs.shape)

csvQ = 'test_imgQ.csv'
imgQs = path2img(csvQ)
print("imgQs:", imgQs.shape)
```

```python
def img2vec(imgs):
    probs = []
    preds = []
    
    for img in imgs:
        img = Image.fromarray(img)
        prob, pred, att = learner.predict(img)
        probs.append(prob)
        preds.append(pred)
    return probs, preds

probEs, predEs = img2vec(imgEs)
probQs, predQs = img2vec(imgQs)
```

### embedding --> middle layer

```python
def get_vecEQs(vecEs, vecQs):
    vecEQs = []
    idxEQs = []
    
    for e in range(len(vecEs)):
        for q in range(len(vecQs)):
            vecEQ = list(vecEs[e]) + list(vecQs[q])
            vecEQs.append(vecEQ)
            idxEQs.append((e, q))
    return vecEQs, idxEQs

''' 
# prob
vecEQs, idxEQs = get_vecEQs(probEs, probQs)
'''

# multi-hot
vecEQs, idxEQs = get_vecEQs(predEs, predQs)
```

```python
import tensorflow as tf
'''
# prob
encoder = tf.keras.models.load_model('../model/gb_prob_encoder_model.keras')
'''

# multi-hot
encoder = tf.keras.models.load_model('../model/gb_multihot_encoder_model.keras')


# middle layer
vecEQs = np.array(vecEQs)
middle_vecEQs = encoder.predict(vecEQs)
```

### middle layer --> score

```python
import joblib

'''
# prob
rf_model = joblib.load('../model/gb_prob_nn_model.pkl')
y_pred = rf_model.predict(middle_vecEQs)
'''

# multi-hot
svm_model = joblib.load('../model/gb_multihot_nn_model.pkl')
y_pred = svm_model.predict(middle_vecEQs)
```

### show top3 & last3

```python
sorted_indices = np.argsort(y_pred.flatten())[::-1]  # good & bad
#sorted_indices = np.argsort(y_pred)[::-1]  # 大 -> 小排序, good & ugly
top_3_indices = sorted_indices[:3]  # top 3
bottom_3_indices = sorted_indices[-3:]  # last 3

top_3_combinations = [(idxEQs[i], y_pred[i]) for i in top_3_indices]
bottom_3_combinations = [(idxEQs[i], y_pred[i]) for i in bottom_3_indices]

print("Top 3 combinations:")
for combo, score in top_3_combinations:
    print(f"Combination: {combo}, Score: {score}")
    display(Image.fromarray(imgEs[combo[0]]))
    display(Image.fromarray(imgQs[combo[1]]))

print("\nBottom 3 combinations:")
for combo, score in bottom_3_combinations:
    print(f"Combination: {combo}, Score: {score}")
    display(Image.fromarray(imgEs[combo[0]]))
    display(Image.fromarray(imgQs[combo[1]]))
```
